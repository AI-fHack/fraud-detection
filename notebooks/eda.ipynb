```python
import pandas as pd

transactions = pd.read_csv("data/transactions.csv", encoding="cp1251")
behavior = pd.read_csv("data/behavior_patterns.csv", encoding="cp1251")
```

```python
# Посмотреть первые 5 строк
print(transactions.head())
print(behavior.head())

# Размер датасета (строки, колонки)
print("Transactions shape:", transactions.shape)
print("Behavior shape:", behavior.shape)

# Типы данных
print(transactions.dtypes)
print(behavior.dtypes)

# Проверка пропусков
print(transactions.isnull().sum())
print(behavior.isnull().sum())

# Процент мошенничества
if 'is_fraud' in transactions.columns:
    fraud_ratio = transactions['is_fraud'].mean()
    print(f"Процент мошеннических транзакций: {fraud_ratio*100:.4f}%")

```

```python
# 'client_id' должен быть в обоих файлах
df = transactions.merge(behavior, on="client_id", how="left")

print("Объединённый датасет shape:", df.shape)
print(df.head())
```
